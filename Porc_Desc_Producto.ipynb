```python
df1 = pd.read_excel("https://drive.usercontent.google.com/download?id=", index_col=0)
```

```python
df2 = df1.drop(['Short Description'], axis=1)
```

```python
df3 = df2['Content'].str.replace('<br>', '_').str.replace('<h3>', '_').str.replace('<b>', '_')
df3.head()
```

```python
df2['Content'].str.contains("Descripción Producto: Porc\\.")
df2.head()
```

```python
df2['PorcDesc'] = df2["Content"].str.contains('Descripción Producto: Porc\\.')
df2.head()
```

```python
# Asegura de que la columna 'Content' sea string antes de limpiar y aplica la limpieza
df2['Content'] = df2['Content'].astype(str).apply(lambda x: re.sub(r'<.*?>', '', x))

df2.to_excel('PorcDesc.xlsx', index=True, engine='xlsxwriter')
```
